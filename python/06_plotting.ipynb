Завдання 1. Побудова графіку динаміки втрат по днях.

````python
pip install matplotlib
````

````python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("processed_losses.csv", encoding="utf-8")


df["Дата"] = pd.to_datetime(df["Дата"])


df = df.sort_values("Дата")


plt.figure(figsize=(12, 6))
plt.plot(df["Дата"], df["Втрати особового складу"], marker='o', linestyle='-')
plt.title("Динаміка втрат по днях")
plt.xlabel("Дата")
plt.ylabel("Втрати особового складу")
plt.grid(True)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
````

Завдання 2. Створення стовпчикової діаграми сумарних втрат по напрямках.

````python
grouped = df.groupby("Напрямок")["Втрати особового складу"].sum()

grouped.plot(kind="bar")
plt.title("Сумарні втрати по напрямках")
plt.ylabel("Втрати особового складу")
plt.xlabel("Напрямок")
plt.grid(axis='y')
plt.show()
````

Завдання 3. Виведення кругової діаграми частки втрат підрозділів.

````python
unit_sum = df.groupby("Підрозділ")["Втрати особового складу"].sum()

unit_sum.plot(kind="pie", autopct="%1.1f%%")
plt.title("Участь підрозділів у бойових діях (за втратами)")
plt.ylabel("")
plt.show()
````

Завдання 4. Побудова двох графіків на одному полотні: по напрямках і по підрозділах.

````python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

df.groupby("Напрямок")["Втрати особового складу"].sum().plot(kind="bar", ax=ax[0])
ax[0].set_title("Втрати по напрямках")

df.groupby("Підрозділ")["Втрати особового складу"].sum().plot(kind="bar", ax=ax[1])
ax[1].set_title("Втрати по підрозділах")

plt.tight_layout()
plt.show()
````

Завдання 5. Збереження одного з графіків у формат `.png`.

````python
plt.plot(df["Підрозділ"], df["Втрати ОВТ"])
plt.title("Втрати ОВТ противника")
plt.savefig("Втрати ОВТ противника.png", dpi=300)
````
